```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/remote_work_productivity.csv')
df.head() #first 5 rows of tha dataset
```

```python
df.describe()#stattical summary
```

```python
df.drop('Employee_ID', inplace=True , axis=1)
```

```python
plt.figure(figsize=(6,7))
ax=sns.countplot(x='Employment_Type',data=df)
ax.bar_label(ax.containers[0],fontsize=14)
plt.show()
```

NUmber of people that work remotely

```python
bars_name=df.groupby(by='Employment_Type').mean().index
plt.figure(figsize=(10,20))
for i,col in enumerate(df.select_dtypes('int').columns) :
    plt.subplot(3,1,i+1) #Fixed: Removed extra space before plt.subplot
    ax=sns.barplot(x=bars_name,y=df.groupby(by='Employment_Type')[col].mean())
    ax.bar_label(ax.containers[0])
    plt.title(col) # Indented this line to include it in the loop
```
